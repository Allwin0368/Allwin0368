```
import csv
with open("energy_dataset.CSV","w",newline="") as f :
      writer = csv.writer(f)
      writer.writerow(["id", "employee count", "company", "energy consumption town"])
      for i in range (1, 501):
           employee_Count = 100+ i
           company = f"Company_{chr (65 + (i %26))}"
           energy =11000+ (1*10)
           writer.writerow([i, employee_Count, company, energy])
```

```
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
samplesize = 100000
uniform = np.random.rand(samplesize)
normal = np.random.randn(samplesize)
plt.hist(uniform, bins=100)
plt.title('rand: uniform')
plt.show()
plt.hist(normal, bins=100)
plt.title('randn: normal')
plt.show()
```
